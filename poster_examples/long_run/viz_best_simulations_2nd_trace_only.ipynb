# Visualization

```python
from brian2 import *
from brian2modelfitting import *

import numpy as np
```

```python
dt = 0.01*ms
defaultclock.dt = dt
```

## Load the Data

```python
voltage1 = genfromtxt('voltage2nA.csv', delimiter=',')
```

```python
inp_cur1 = genfromtxt('inp_cur2nA.csv', delimiter=',')
```

```python
out_spikes1 = genfromtxt('out_spikes2nA.csv', delimiter=',')
```

```python
spikes1 = genfromtxt('spikes2_hh.csv', delimiter=',')
fits = genfromtxt('fits_hh.csv', delimiter=',')
# fits_vt = genfromtxt('fits_vt_hh.csv', delimiter=',')
```

```python
t0 = np.arange(0, len(fits[0]))*dt/ms
```

```python
plt.style.use('seaborn-poster')
```

```python
plt.figure(figsize=(30,5))
# plot(t0, fits[0].transpose());
vlines(spikes1, fits[1][np.int_(np.round(spikes1/dt*ms))]-75, fits[1][np.int_(np.round(spikes1/dt*ms))]+60, 'r');
plot(t0, voltage1[:5000000]);
xlim(9000, 14000);
```

```python
d0, d1 = 500000, 650000
spikes_temp = spikes1[spikes1<d1/100]
spikes_temp = spikes_temp[spikes_temp>d0/100]

out_spikes_temp = out_spikes1[out_spikes1<d1/100]
out_spikes_temp = out_spikes_temp[out_spikes_temp>d0/100]


fig, ax = plt.subplots(nrows=3, figsize=(30, 8), sharex=True)
ax[0].plot(t0[d0:d1], inp_cur1[d0:d1]);

ax[1].plot(t0[d0:d1], voltage1[d0:d1]);
ax[1].vlines(out_spikes_temp, voltage1[d0:d1][np.int_(np.round(out_spikes_temp))]-40, voltage1[d0:d1][np.int_(np.round(out_spikes_temp))],linestyles='dotted');
ax[1].set_ylim(-120,80)
ax[1].set_xlim(d0/100, d1/100)

ax[2].plot(t0[d0:d1], fits[1][d0:d1].transpose(), 'g');
ax[2].vlines(spikes_temp, fits[1][d0:d1][np.int_(np.round(spikes_temp))], fits[1][d0:d1][np.int_(np.round(spikes_temp))]+10, 'g');
ax[2].vlines(spikes_temp, fits[1][d0:d1][np.int_(np.round(spikes_temp))]+10, fits[1][d0:d1][np.int_(np.round(spikes_temp))]+20, linestyles='dotted');
ax[2].set_ylim(-7,12);
```
